```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Bar, Line, Pie
import os
pd.set_option("display.max_colwidth", -1)
```

```python
fpath = "E:/code/data_analysis/pandas_data_analysis/data/blog_access_log"
df_list = []
for file_ in os.listdir(fpath):
    file_data = pd.read_csv(f"{fpath}/{file_}",sep = " ", header=None, error_bad_lines = False)
    df_list.append(file_data)
df = pd.concat(df_list)
```

```python
df.head()
```

```python
df = df[[0,3,6,9]].copy()
df.head()
```

```python
df[3] = df[3].str[1:]
df.head()
```

```python
df.columns = ["ip", "stime", "status", "client"]
df.head()
```

```python
import re
df['contain'] = df["client"].str.lower().str.contains("spider")
df.head()
```

```python
df_spider = df['contain'].value_counts()
```

```python
bar=(
    Bar()
    .add_xaxis([str(x) for x in df_spider.index])
    .add_yaxis("是否siper", df_spider.values.tolist())
    .set_global_opts(title_opts = opts.TitleOpts(title="爬虫访问量占比"))
)
bar.render_notebook()
```

```python
df_status = df["status"].value_counts()
df_status
```

```python
a = list(zip(df_status.index, df_status))
```

```python
pie = (
    Pie()
    .add("状态码比例", a)
    .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}:{c}'))
)
pie.render_notebook()
```

```python
df.head()
```

```python
df.set_index(pd.to_datetime(df["stime"], format="%d/%b/%Y:%H:%M:%S"),inplace=True)
```

```python
df.sort_index()
df.head()
```

```python
df.index
```

```python
df_pvuv = df.resample("D")["ip"].agg(pv=np.size, uv = pd.Series.nunique)
df_pvuv.head()
```

```python
line= (
    Line()
    .add_xaxis(df_pvuv.index.to_list())
    .add_yaxis("pv",df_pvuv['pv'].to_list())
    .add_yaxis("uv",df_pvuv['uv'].to_list())
    .set_global_opts(title_opts= opts.TitleOpts(title="PVUV数据对比"),tooltip_opts= opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"))
)
line.render_notebook()
```
