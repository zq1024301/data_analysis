```python
import pandas as pd
```

```python
grade_data = pd.read_excel("E:/code/data_analysis/pandas_data_analysis/data/12_学生成绩表.xlsx")
grade_data.head()
```

```python
information_data = pd.read_excel("E:/code/data_analysis/pandas_data_analysis/data/13_学生信息表.xlsx")
information_data.head()
```

```python
new_data = information_data.loc[: , "学号":"性别"]
new_data.head()
```

```python
a = pd.merge(left=grade_data,right=new_data,on='学号')
a.head()
```

```python
old_index = a.columns.to_list()
old_index
```

```python
new_insert_index = a.columns.to_list()[-2:]   
new_insert_index
```

```python
for index_ in new_insert_index[::-1]:
    old_index.remove(index_)
    old_index.insert(old_index.index('学号')+1, index_)
old_index
```

```python
a.reindex(columns = old_index)
```
