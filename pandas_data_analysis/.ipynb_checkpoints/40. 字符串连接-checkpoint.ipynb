```python
import pandas as pd
```

```python
fpath = "E:/code/data_analysis/pandas_data_analysis/data/4_beijing_tianqi_2018.csv"
df = pd.read_csv(fpath)
df.head(3)
```

```python
df["bWendu"] = df["bWendu"].str.replace("℃", '').astype("int32")
df["yWendu"] = df["yWendu"].str.replace("℃", '').astype("int32")
df.head(3)
```

```python
df.set_index(pd.to_datetime(df["ymd"]),inplace=True)
df.head(3)
```

```python
df.index.month
```

```python
result = (df.groupby(df.index.month)
         .agg(最高温度= ("bWendu","max"),
              风向 = ("fengxiang", lambda x: ",".join(x.unique())))
         .reset_index()
             )
```

```python
result
```

```python
def abc(x):
    return pd.Series({
        "最高温度":x["bWendu"].max(),
        "风向":",".join(x["fengxiang"].unique())
    })

result1 = df \
        .groupby(df.index.month) \
        .apply(abc) \
        .reset_index() \
        .rename(columns={"ymd":"月份"})
```
