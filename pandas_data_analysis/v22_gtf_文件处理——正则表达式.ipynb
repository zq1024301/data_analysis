```python
import pandas as pd
import re
```

```python
file_path = "C:/Users/sunny/Desktop/ABC/gencode.v22.annotation.gtf"
gene_annotation = pd.read_csv(file_path, skiprows=5, header=None, sep= "\t")
gene_annotation.head()
```

```python
import time
starttime = time.time()
file_path = "C:/Users/sunny/Desktop/ABC/gencode.v22.annotation.gtf"
gene_id_list = []
gene_type_list = []
gene_name_list = []
f = open(file_path, 'r')
for line in f.readlines():
    if line.startswith("#"):
        continue
    gene_id = re.search(r'gene_id ".+?"', line).group()[11:-1]
    gene_type = re.search(r'gene_type ".+?"', line).group()[11:-1]
    gene_name = re.search(r'gene_name ".+?"', line).group()[11:-1]
    gene_name_list.append(gene_name)
    gene_id_list.append(gene_id)
    gene_type_list.append(gene_type)
gene_type_df = pd.DataFrame(gene_type_list, columns=["gene_type"])
gene_id_df = pd.DataFrame(gene_id_list, columns=["gene_id"])
gene_name_df = pd.DataFrame(gene_name_list, columns=["gene_name"])
df1 = pd.concat([gene_name_df, gene_type_df, gene_id_df], axis=1)
endtime = time.time()
print(endtime-starttime)
```
